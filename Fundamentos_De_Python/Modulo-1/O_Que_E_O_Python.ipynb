# O que é Python

Python é uma linguagem de programação de alto nível, interpretada, de propósito geral e multiparadigma, criada por Guido van Rossum e lançada em 1991. Ela foi projetada para ser simples de ler, escrever e manter, tornando-se uma das linguagens mais populares do mundo.

## Principais características

* **Sintaxe simples e legível:** utiliza uma escrita próxima da linguagem humana.
* **Interpretada:** o código é executado por um interpretador, sem necessidade de compilação prévia na maioria dos casos.
* **Multiparadigma:** suporta programação procedural, orientada a objetos e funcional.
* **Multiplataforma:** funciona em Windows, Linux e macOS.
* **Código aberto:** é gratuita e mantida por uma grande comunidade.
* **Grande ecossistema:** possui milhares de bibliotecas prontas para diversas finalidades.

## Como o Python funciona

Quando você escreve um programa em Python, o interpretador lê o código e o executa instrução por instrução. Por exemplo:

```python
print("Olá, mundo!")
```

Saída:

```text
Olá, mundo!
```

Nesse exemplo, a função `print()` exibe uma mensagem no terminal.

## Exemplo prático



