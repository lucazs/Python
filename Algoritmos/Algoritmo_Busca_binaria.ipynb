## 📘 O que é a Pesquisa Binária?

A pesquisa binária é um algoritmo de busca altamente eficiente, mas que possui uma condição obrigatória: a lista de elementos precisa estar estritamente ordenada.

Diferente da busca simples (ou sequencial), onde você testa elemento por elemento de forma linear (o que pode levar $N$ etapas para uma lista de tamanho $N$), a pesquisa binária adota a estratégia de "dividir para conquistar":

- Ela faz um "chute" exatamente no elemento do meio da lista.
- Se o palpite for maior ou menor do que o item procurado, metade das possibilidades restantes é descartada imediatamente.
- Esse processo se repete consecutivamente na metade que sobrou até que o item seja localizado ou a busca se esgote.



```python
# Jogo de adivinhação
import random
limite_baixo = 0
limite_alto = 10

escolha = random.randint(limite_baixo, limite_alto)

while True:

    N = int(input("Escolha um número de 1 até 10."))
    print(N)

    if escolha == N:
        print("Acertou")
        break
    else:
        if escolha > N:
            print("Baixo Demais...")
        else:
            print("Alto Demais...")
```
