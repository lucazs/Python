# Notação Big O

## Resumo do conteúdo

A live conclui os principais conceitos do **Capítulo 1**, concentrando-se na análise da eficiência dos algoritmos por meio da **Notação Big O**.

A ideia central é que não devemos avaliar um algoritmo apenas pelo tempo que ele leva para executar uma tarefa em um computador específico. O mais importante é entender **como o número de operações cresce conforme aumenta a quantidade de dados de entrada (`N`)**.

### Pesquisa simples vs. pesquisa binária

A **pesquisa simples** verifica os elementos um por um. Por isso, seu tempo de execução cresce proporcionalmente ao tamanho da lista, resultando em uma complexidade **O(N)**.

Já a **pesquisa binária** divide repetidamente o espaço de busca pela metade. Dessa forma, mesmo com uma quantidade muito grande de elementos, o número de operações cresce lentamente, apresentando complexidade **O(log N)**.

Por exemplo, em uma lista com aproximadamente **1 bilhão de elementos**, uma pesquisa binária precisa de cerca de **30 tentativas**, enquanto uma pesquisa simples pode precisar verificar até 1 bilhão de elementos.

> **Ideia principal:** quanto mais eficiente for a redução do espaço de busca, menor será o crescimento do número de operações.

## Principais complexidades apresentadas

Da mais eficiente para a menos eficiente:

| Complexidade   | Tipo         | Exemplo                          |
| -------------- | ------------ | -------------------------------- |
| **O(log N)**   | Logarítmica  | Pesquisa binária                 |
| **O(N)**       | Linear       | Pesquisa simples                 |
| **O(N log N)** | Quase linear | Quicksort                        |
| **O(N²)**      | Quadrática   | Ordenação por seleção            |
| **O(N!)**      | Fatorial     | Força bruta no Caixeiro-Viajante |

A diferença entre essas complexidades se torna cada vez mais significativa à medida que `N` aumenta.

## O Problema do Caixeiro-Viajante

O problema consiste em encontrar a **menor rota possível para visitar várias cidades**, passando por cada uma delas e retornando ao ponto inicial.

Uma solução por força bruta precisa testar todas as possíveis combinações de rotas. Como o número de possibilidades cresce de forma **fatorial — O(N!)** —, o problema rapidamente se torna impraticável para grandes quantidades de cidades.

Esse exemplo demonstra uma importante lição:

> Um algoritmo que funciona bem com poucos dados pode se tornar completamente inviável quando a quantidade de dados aumenta.

## O pior caso

A Notação Big O normalmente analisa o **pior caso possível**.

Por exemplo, em uma busca linear:

* Melhor caso: o item está na primeira posição;
* Caso médio: o item está em alguma posição intermediária;
* Pior caso: o item está no final da lista ou não existe.

A análise do pior caso permite estabelecer uma **garantia de desempenho** para o algoritmo.

## Exercícios do capítulo

A live também discute situações práticas envolvendo uma agenda telefônica:

* **Encontrar o telefone a partir do nome:** `O(log N)`, se os nomes estiverem ordenados;
* **Encontrar o dono a partir do número:** `O(N)`, pois pode ser necessário verificar toda a agenda;
* **Ler todos os telefones:** `O(N)`;
* **Ler os telefones de todas as pessoas cujo nome começa com A:** continua sendo `O(N)` no pior caso.

Esse último exemplo é importante porque mostra que **filtrar os dados não necessariamente reduz a complexidade do algoritmo**. Se for necessário verificar todos os registros para descobrir quais atendem à condição, o algoritmo continua tendo comportamento linear.
